```python
coal = df2["CEA.DGR.COL"]
year= df2.yyyymmdd
```

```python
df2['Year'] = df2['yyyymmdd'] // 100
df2['Month'] = df2['yyyymmdd'] // 100
df2['Day'] = df2['yyyymmdd'] // 100
```

```python
df2['Financial_Year'] = df2.apply(
    lambda row: f"{row['Year']}-{row['Year'] + 1}" if row['Month'] >= 4 else f"{row['Year'] - 1}-{row['Year']}",
    axis=1
)
ap = df2.groupby('Financial_Year')["CEA.DGR.COL"].sum().reset_index()
ap.columns = ['Financial_Year', 'Coal_Electricity_Production']

print(ap)
```

```python
# Remove rows from index 0 to 10 (inclusive)
ap=ap.drop(index=range(0, 16))

# Reset the index if needed
ap=ap.reset_index(drop=True)
print(ap)
```

```python
ap=ap.drop(index=10)
ap
```

```python
grouped_data = ap.groupby('Financial_Year')['Coal_Electricity_Production'].sum()
plt.figure(figsize=(10, 6))
plt.plot(grouped_data.index, grouped_data.values)
plt.xlabel('Financial Year')
plt.ylabel('Coal Electricity Production (Million kWh)')
plt.title('Coal Electricity Production per Financial Year')
plt.xticks(rotation=75, ha='right')
plt.tight_layout()
plt.show()
```
